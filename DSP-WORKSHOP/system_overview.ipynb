# System Overview

In this workshop, you are not building an entire system from scratch.

Instead, you are building **small modules** that plug into a larger audio system.

Throughout this diagram you can see different blocks being connected to each other. Each block is called a **module.** You can think of a module like a function (or method) in other programming languages like Python or C/C++ where the look at the syntax below kinda like passing arguments into a function. ## Module Instantiation

To use a module inside another file, we **instantiate** it. 

**General Syntax:**

```systemverilog
module_name instance_name (
    .port1(signal1),
    .port2(signal2)
);
```

### Overriding Parameters (like constants)
Modules can also have parameters, which are like configurable constants.

We can override them when instantiating the module:

**General Syntax:**

```systemverilog
module_name #(
    .CONSTANT_A(20'sd8),
    .CONSTANT_B(-20'sd16)
) instance_name (
    .port1(signal1),
    .port2(signal2)
);
```
