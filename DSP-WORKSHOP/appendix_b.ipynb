# Appendix B: Deeper Explanation of Ring Modulation

## Understanding AM Through Radio

In general, modulation takes a raw signal (like sound) and transforms it into a form that can travel long distances and be handled by electronic systems. Amplitude modulation is just one technique for modulation and one of the simplest to implement.

Every radio station has a modulator that performs the same process we used in this workshop to transmit sound. Every radio receiver has a demodulator that reverses this process to recover the original audio we hear. The signal $c[n]$ is called the carrier signal because it carries the sound from the station to us.

### Why do we multiply by a cosine (or sine)?

There are many reasons we multiply by a cosine (or sine), but one of them is to enable radio channels. There are also many more technical reasons for this but those are usually covered in upper-year courses.

Radio stations often have names like 99.9 Virgin Radio Toronto or 100.5 Classic Rock. These numbers represent the frequency bands each station occupies. Each station transmits by multiplying its raw audio signal with a cosine (or sine) at a specific frequency, which determines its channel.

If radio stations transmitted only raw signals, everything would overlap and interfere. Instead, each station uses a different carrier frequency so signals can coexist without interference.

A cosine signal can be written as:
$$c[n] = \cos(2\pi f n)$$
where $f$ is the frequency of the carrier.
