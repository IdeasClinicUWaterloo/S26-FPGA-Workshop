# SystemVerilog Reference Sheet

This contains all the syntax you’ll need for this workshop. If you’re unsure what something means, check this [guide](./systemverilog_guide/table_contents.ipynb).


## Types of Blocks
|  | Syntax | Notes |
| :---:     | :---   | :---:     |
| Continuous | `assign x = a + 1;` | Always active, no clock |
| Sequential | `always_ff @(posedge clk)` | Clocked logic |

## Assignment Operators
|  | Syntax | Notes |
| :---:    | :---  | :---:    |
| Non-blocking assignment | `a <= b;` | Scheduled update (use in `always_ff`) |


## Variables and Arrays
|  | Syntax | 
| :---:    | :---    |
| Variable declaration | `logic [15:0] x;` |
| Array declaration | `logic [15:0] arr [0:511];` |

## If-Else Statement
|  | Syntax | 
| :---:   | :---    |
| If statement | <code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end </code>|
| If-else statement | <code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end <br> else begin <br> &nbsp; result <= a - b; <br> end </code> |
| Else-if | <code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end <br> else if (a == 2) begin <br> &nbsp; result <= a - b; <br> end </code> |

## Module
|  | Syntax | Notes |
| :---:| :---| :---:|
| Module definition | <code>module my_module ( <br> &nbsp; input logic clk, <br> &nbsp; input logic [15:0] a, <br> &nbsp; output logic [15:0] y <br> ); <br><br> // logic here <br><br> endmodule </code> | Defines a module |
| Module instantiation | <code>my_module u1 ( <br> &nbsp; .clk(clk), <br> &nbsp; .a(a), <br> &nbsp; .y(y) <br> ); </code> | Connects modules |

## Common Operations
|  | Syntax | Notes |
| :---:| :---| :---:|
| Scaling | `y = x >>> 1;` | Divide by 2 |
