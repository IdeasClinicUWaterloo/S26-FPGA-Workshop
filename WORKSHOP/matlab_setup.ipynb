# MATLAB Set-up

We will walk you through how to set up a MATLAB account and create a script file.

## MATLAB Registration
1. Go to the University of Waterloo’s [MATLAB Portal](https://www.mathworks.com/academia/tah-portal/university-of-waterloo-31483447.html).
2. Click **Sign in to get started**.
3. If you do not already have an account, click  **Create an account**.
    - Enter your UWaterloo e-mail
    - Create a password
    - Click **Next**
4. Verify your account using the e-mail address you registered with. 
5. Fill in your personal details and click **Next**

Your MathWorks account is now created.



## Use MATLAB online
1. Sign into [MathWorks](https://www.mathworks.com/academia/tah-portal/university-of-waterloo-31483447.html) and click **Get started**
2. Click **Open MATLAB Online** to use the software online.
3. Click **New**, then **Script** to create a new `.m` file 
4. When you click run for the first time:
    - MATLAB will prompt you to save the file
    - You can either: click **Save** or give the file a name and save it



Now you can go back to the [filter design section](matlab_filter.ipynb) and run the provided MATLAB scripts to generate the filter coefficients for your design.
