# Appendix: Test Benches

This appendix covers how to write and run a Verilog test bench. Test benches are used to verify your circuits in software _before_ programming them onto a real FPGA.

## Testing digital_example

A test bench is a Verilog module with no inputs and no outputs.
Its only job is to create input signals, feed them into your design (the _design under test_, or DUT),
and check that the outputs match what you expect.

Let's test the module from the [digital logic section](./digital_logic.ipynb):

```verilog
module digital_example(
    input a,
    input b,
    input c,
    output out
);
    logic gate_1, gate_2, gate_3;
  
    assign gate_1 = a & b;
    assign gate_2 = ~c;
    assign gate_3 = gate_1 ~| gate_2;
    assign out = gate_3 ^ gate_2;

endmodule
```

Here is a test bench that checks every row of its truth table:

```verilog
module test_digital_example;
    logic [3:0] test_vector [7:0];
    logic a_test, b_test, c_test, out;
    logic tb_out;

    // Instantiate the design under test
    digital_example de_inst(
        .a(a_test),
        .b(b_test),
        .c(c_test),
        .out(out)
    );

    initial begin
        // Each vector is 4 bits: {a, b, c, expected_out}
        test_vector = '{
            4'b0001,
            4'b0011,
            4'b0101,
            4'b0111,
            4'b1001,
            4'b1011,
            4'b1101,
            4'b1110
        };

        for (int i = 0; i < 8; i++) begin
            {a_test, b_test, c_test, tb_out} = test_vector[i];
            #1; // wait one time step for outputs to settle
            assert (out == tb_out);
        end
        $display("All tests passed!");
        $finish;
    end
endmodule
```


## Running the Test Bench

You can run this with [Verilator](https://veripool.org/verilator/) (a free Verilog simulator):

```bash
verilator test_digital_example.sv --binary --timing
./obj_dir/Vtest_digital_example
```

Expected output:

```
All tests passed!
```

If any assertion fails, Verilator will print which line failed.
This allows you to determine why specific inputs might have given you outputs you did not expect.
