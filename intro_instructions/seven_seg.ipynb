# Seven Segment Display

## Seven-Segment Displays

![Seven Segment Display](./assets/seven_segment.png)

To drive the seven-segment display, a seven-digit binary number, formatted as "gfedcba", is assigned to the display, using the following syntax.

```verilog
assign HEX0 = 7'b1000000;
```

> [!NOTE]
> The letters refer to the order of the bits, not to the method of writing down numbers in hexadecimal.
> "a" marks the position of the Least Significant bit and "g" marks the most significant bit.

This code corresponds to the seven-segment display closest to the right side of the board. The display is active low, so to turn a segment on, it should be assigned a 0. So, to display the digit 7 we need to set segments 'a', 'b', and 'c' to low. To display this number on the rightmost display the instruction would be,

```verilog
assign HEX0 = 7'b1111000;
```

**NOTE regarding the use of quotation marks. For multi-bit values (e.g., `logic [6:0]`), double quotation marks with a width prefix should be used, such as `7'b1000000` for driving the seven-segment display. For individual logic states (e.g., `logic`), single-bit literals like `1'b0` or `1'b1` are used.

## Always Block

So far we have used `assign` for simple one-line logic. When you need multi-line logic, you use an `always_comb` block instead.

Think of `always_comb` as a stand-in for a **black box** that can contain as many gates as you need. Inside the black box you can write any number of lines of logic; the tools will figure out the actual gate circuit. Every time any input to the block changes, the entire block re-evaluates and the outputs update.

<!-- TODO: insert diagram showing a black box labelled always_comb, zoomed in to reveal random gates inside -->

> **Combinational logic** is logic where the outputs depend only on the current inputs, with no memory of past values. An `always_comb` block describes combinational logic: as soon as the inputs change, the outputs are recomputed. All of the digital gates we have presented are combinational in nature. (Later you will learn about `always_ff`, which describes sequential logic that _does_ remember past values.)

```verilog
always_comb begin
    // your logic here
end
```

Variables to be used within `always` blocks are declared as `logic` type outside of the block itself.

## Multiplexers using if, else, and case

A multiplexer is another name for a digital switch.
Switches are important because they allow the activation of parts of our circuit based on different operating conditions.
This _switching action_ is represented in Verilog using two constructs borrowed from programming languages, the `if`-`else if`-`else` statements, and the `case` statement.

> Verilog is a Hardware Description Language (HDL), not a programming language.
> Even though it borrows a lot of keywords and is "compiled" into a bitstream using similar methods,
> the key difference is that Verilog describes a static **layout** and NOT a sequential set of instructions.
> This layout is realized using gates which are _inferred_ from the description of behaviour which Verilog provides.

```verilog
if (condition)
    // do something
else
    // do something else
```

For instance, the circuit described by

```verilog
if (sel = 1'b0)
    option_a <= out;
else
    option_b <= out;
```

would be inferred to look like:

![A trapezoid representing the multiplexer element connecting option_a, option_b to output](./assets/multiplexer.png)

`sel`, standing for "select", is a common name to refer to the value used to pick what the value of the multiplexer's output will be.

### The case statement

The case statement provides the same functionality, but uses a list of specific values which select may take. The list of values is converted into a chain or tree of individual multiplexer switches.

```verilog
case (signal)
    2'b00: 
    2'b01:
    2'b10:
    2'b11:
    default:
endcase
```

The `default` case handles any value not explicitly listed. You will use `case` statements in the activity below to map switch values to seven-segment display patterns.

## Example Code

This is a simple script that checks the value of two switches, and if they are high, displays a 1 on the seven-segment display. Otherwise, the display shows 0. You can use this as a reference to understand the syntax.

```verilog
module SegExample(
    input  [1:0] SW,
    output logic [6:0] HEX0  // 7-seg display (active-low)
);

// SIGNAL declarations (not used in this example)

// Always block
always_comb begin  // the block will run anytime there is a change to the SW inputs
    if (SW[0] == 1'b1 && SW[1] == 1'b1)
        HEX0 = 7'b1111001; // Display "1"
    else
        HEX0 = 7'b1000000; // Display "0"
end
// Always block end

endmodule
```

## Instructions

The goal of this activity is to drive the seven-segment display using on-board switches. Each switch being flipped should correspond to a number being displayed on the board, e.g. when SW3 is flipped; a 3 is displayed on the seven-segment display.

To write the body of this Verilog code, start by declaring a signal to store the binary number corresponding to the active switches. Then, create an `always_comb` block that monitors the state of all input switches.

Use a series of `if` statements (or a `case` statement) to determine which switch is turned on, giving priority to the one with the lowest index. When a switch is found to be active, assign its index value (in binary) to the signal you created. If no switches are on, assign a default value to indicate "no input."

Finally, ensure the output is assigned to the 7-segment display port HEX0, so the number appears on the screen.

> [!IMPORTANT]
> Even though your code will use signals `SW[3:0]` (sliding switch number three all the way to switch zero, **four switches in total**), our skeleton code will reassign those signals to `SW[7:4]` instead.
>
> **This means that you should test sliding switch 7 (`SW[7]`) on the physical board to see what happens when `SW[3]` is changed in your digital circuit.**
