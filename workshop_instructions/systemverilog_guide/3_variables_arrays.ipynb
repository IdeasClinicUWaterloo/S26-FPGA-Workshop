## 3. Variables and Arrays

### Variables

Examples:

```systemverilog
logic a;                  // holds 1 bit (0 or 1)
logic [7:0] b;            // holds 8 bits (0 to 255)
logic signed [7:0] c;     // holds 8 bits and can be negative (-128 to 127)
```

#### How many bits do you need 
The number of bits determines how large a value you can store.

With $N$ bits, you can you can represent values from $0$ to $2^{N} - 1$

To store the number 25:
- 4 bits: too small (max number it can hold is 15)
- 5 bits: just right (max number it can hold is 31)

### Array

Example:

```systemverilog
logic [15:0] arr [0:7];
```

This is means:
- The array that has 8 elements
- Each element is 16 bits wide

### Literals

A literal is a fixed value written directly in the code to represent constant data, such as numbers.

For example, to write a binary 1 or 0 (true/false), we use:
- `1'b1` for 1 (1-bit wide, binary)
- `1'b0` for 0 (1-bit wide, binary)

Literals follow this format:
`<width>'<base><value>`

For example:
- 1-bit binary value of 1: `1'b1`
- 8-bit decimal value of 10: `8'd10`

In the workshop, the filter coefficients are 20-bit signed values, meaning they can be positive or negative.  
For example:
- +250 (20-bit signed decimal): `20'sd250`
- -250: `-20'sd250`
