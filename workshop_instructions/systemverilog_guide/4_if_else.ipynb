## 4. If-Else Statement

Used inside `always_ff` blocks.

General Syntax:
```systemverilog
if (<condition>) begin
    // statement 1
end 
else begin
    // statement 2
end
```
Example:
```systemverilog
if (a == 1) begin
    result <= a + b;
end 
else if (a == 2) begin
    result <= a - b;
end
else begin
    result <= a * b;
end
```
