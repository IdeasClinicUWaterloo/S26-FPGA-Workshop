## Variables and Arrays

<table style="margin-left: 0; margin-right: auto;">
  <tr>
    <th></th>
    <th>Syntax</th>
    <th>Notes</th>
  </tr>
  <tr>
    <td>Variable declaration</td>
    <td><code>logic [15:0] x;</code></td>
    <td>16 bit wide variable that can hold unsigned decicamls (0 to 65535)</td>
  </tr>
  <tr>
    <td>Signed variable declaration</td>
    <td><code>logic signed [15:0] x;</code></td>
    <td>16 bit wide variable that can hold signed decicamls (−32768 to 32767)</td>
  </tr>
  <tr>
    <td>Array declaration</td>
    <td><code>logic [15:0] arr [0:511];</code></td>
    <td>Array of 512 elements, each a 16-bit wide variable that can represent unsigned values (0 to 65535)</td>
  </tr>
</table>

## If-Else Statement

<table style="margin-left: 0; margin-right: auto;">
  <tr>
    <th></th>
    <th>Syntax</th>
  </tr>
  <tr>
    <td>If statement</td>
    <td><code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end </code></td>
  </tr>
  <tr>
    <td>If-else statement</td>
    <td><code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end <br> else begin <br> &nbsp; result <= a - b; <br> end</code></td>
  </tr>
  <tr>
    <td>Else-if</td>
    <td><code>if (a == 1) begin <br> &nbsp; result <= a + b; <br> end <br> else if (a == 2) begin <br> &nbsp; result <= a - b; <br> end</code></td>
  </tr>
</table>
