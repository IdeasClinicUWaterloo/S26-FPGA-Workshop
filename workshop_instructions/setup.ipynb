# Set-up

Before we start, you need to download and open the skeleton project in Quartus for this workshop. 

This project contains pre-written files so you don't have to build everything from scratch. 

---

## Download and Open the Project

1. Download the project

Download and unzip the skeleton project ZIP file [here](https://github.com/IdeasClinicUWaterloo/UpdatedFPGAWorkshop/blob/feat_dsp_workshop/quartus_skeleton.zip). 

**Note:**

GitHub may say the file is too large to preview. This is normal—just ignore it and download the file.



2. Open Quartus

Launch **Quartus Prime** on your computer.

3. Open the project

- Click **File**, then **Open Project**
- Navigate to the unzipped folder
- Select the `.qpf` file



4. Switch to `Files` view

In Quartus:
- Find the **Project Navigator** panel (usually on the left)
- Change the dropdown from **Hierarchy** to **Files**

You should now see a list of all project files.



---

## Project Structure

You've probably noticed that there are **a lot of files**, this is normal. Let's break it down so you don't get too overwhelmed. 

### What you will be doing 

In this workshop, you will:
- Write your own modules for different sound effects
- Add them inside `audio_processing.sv`

In short, you are building small components that plug into a larger audio system.

### Files you will work with
- `audio_processing.sv`
    - This is your main file
    - You will add your modules and connect them here
- Your own modules (e.g. `echo.sv`, `robotic.sv`, etc.)

### Files you will NOT change

This project has many supporting files running in the background.

- DAC-I2C folder: Handles audio input/output (microphone and speaker/headphones)
- FFT-MAGNITUDE folder: Processes audio signals for frequency analysis
    - FFT: converts signal to frequency domain
    - Magnitude: computes strength of each frequency
- HDMI folder: Handles displaying the output on screen

You do **NOT** need to modify any of these.

Focus only on your sound effect modules and `audio_processing.sv`. Everything else is already handled for you.

---

## **Important: Change Compiler Settings**

To avoid long compile times, we will disable an advanced optimization setting.

1. In Quartus:
- Click **Assignments**, and then **Settings**

2. In the Settings window:
- Click **Compiler Settings** (usually on the left sidebar)
- Then click **Advanced Settings (Fitter)**

3. Find this option: 
-  **Advanced Physical Optimization**
4. Set it to: 
- **Off**
5. Click **OK** to save your changes

Don’t worry if you don’t understand this setting, we’re just doing it to make things faster.

## Initial Compile and Flash

1. Compile the project in Quartus
2. Program (flash) the FPGA onto your board

### What you should see and hear
After programming the board:
- Plug in headphones or a speaker
- You should hear audio
- You should see the FFT of the audio displayed on the screen


### `audio_processing.sv`

#### Where this module is: 


#### What this module does: 

This is the main module you will modify during the workshop. 

Currently it is just a passthrough where:
- The input audio is sent directly to the output
- No effects are being applied yet

You will replace the passthrough with sound effect modules.

`audio_processing.sv` currently looks like this:

```systemverilog
module audio_processing (
    input logic clk_50,
    input logic reset,
    input logic [9:0] sw,
    input logic signed [15:0] in_audio,
    input logic in_valid,
    output logic signed [15:0] out_audio,
    output logic out_ready
);

    assign out_audio = in_audio;
    assign out_ready = in_valid;

endmodule
```
